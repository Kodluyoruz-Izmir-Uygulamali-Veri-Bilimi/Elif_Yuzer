# Soru 5:

1.sınıf, 2.sınıf ve 3.sınıf yolcuların ölüm oranlarını karşılaştırınız. Bir veya bir kaç cümle ile sonucu <b>yorumlayınız.</b>


```python

# bu alana python kodunu yazıp çıktısı görülecek şekilde çalıştırınız

df_titanic
df_titanic[["Survived", "Pclass"]]
firstc= df_titanic[(df_titanic.Survived==0) & (df_titanic.Pclass==1)]
secondc = df_titanic[(df_titanic.Survived==0) & (df_titanic.Pclass==2)]
thirdc = df_titanic[(df_titanic.Survived==0) & (df_titanic.Pclass==3)]
query1 = len(firstc[firstc["Survived"]==0])
print("1.sınıf ölen sayısı: ",query1)
query2 = len(secondc[secondc["Survived"]==0])
print("2.sınıf ölen sayısı: ", query2)
query3 = len(thirdc[thirdc["Survived"]==0])
print("3.sınıf ölen sayısı: ", query3)
sumfirst= df_titanic[(df_titanic.Pclass==1)]
sumfirst = len(sumfirst)
print("1.sınıf toplam yolcu: ", sumfirst)
sumsecond= df_titanic[(df_titanic.Pclass==2)]
sumsecond = len(sumsecond)
print("2.sınıf toplam yolcu: ", sumsecond)
sumthird= df_titanic[(df_titanic.Pclass==3)]
sumthird = len(sumthird)
print("3.sınıf toplam yolcu: ", sumthird)
birincisınıf=(query1/sumfirst)*100
print("1.sınıf ölüm oranı:%",birincisınıf)
ikincisınıf=(query2/sumsecond)*100
print("2.sınıf ölüm oranı:%",ikincisınıf)
ucuncusınıf=(query1/sumthird)*100
print("3.sınıf ölüm oranı:%",ucuncusınıf)
if    birincisınıf>ikincisınıf:
    print ("1.sınıfın ölüm oranı 2.sınıftan fazladır")
elif birincisınıf<ikincisınıf:
    print ("1.sınıfın ölüm oranı 2.sınıftan azdır")
else:
    print ("Ölüm oranları eşittir")
if    birincisınıf>ucuncusınıf:
    print ("1.sınıfın ölüm oranı 3.sınıftan fazladır")
elif birincisınıf<ucuncusınıf:
    print ("1.sınıfın ölüm oranı 3.sınıftan azdır")
else:
    print ("Ölüm oranları eşittir")
if    ucuncusınıf>ikincisınıf:
    print ("3.sınıfın ölüm oranı 2.sınıftan fazladır")
elif ucuncusınıf<ikincisınıf:
    print ("3.sınıfın ölüm oranı 2.sınıftan azdır")
else:
    print ("Ölüm oranları eşittir")
```

# Soru 6:

Kazada ölen kadınların oranı ile erkeklerin oranını karşılaştırınız. Bir veya bir kaç cümle ile sonucu <b>yorumlayınız.</b>

```python
# bu alana python kodunu yazıp çıktısı görülecek şekilde çalıştırınız. 
df_titanic[["Survived", "Sex"]]
deadfemale=df_titanic[(df_titanic.Survived==0) & (df_titanic.Sex=="female")]
queryfemale= len(deadfemale[deadfemale["Survived"]==0])
print("Ölen kadın sayısı: ",queryfemale)
deadmale=df_titanic[(df_titanic.Survived==0) & (df_titanic.Sex=="male")]
querymale= len(deadmale[deadmale["Survived"]==0])
print("Ölen erkek sayısı: ",querymale)
allfm=df_titanic[(df_titanic.Sex=="female")]
queryfm=len(allfm[allfm["Sex"]=="female"])
print("Toplam Kadın yolcu: ",queryfm)
allm=df_titanic[(df_titanic.Sex=="male")]
querym=len(allm[allm["Sex"]=="male"])
print("Toplam Erkek yolcu: ",querym)
kadınoranı=(queryfemale/queryfm)*100
erkekoranı=(querymale/querym)*100
print("Ölen kadın yolcu oranı:% ",kadınoranı)
print("Ölen erkek yolcu oranı:% ",erkekoranı)
if kadınoranı<erkekoranı:
    print("Ölen erkek yolcu oranı fazladır")
elif erkekoranı<kadınoranı:
    print("ölen kadın yolcu oranı fazladır")
else:
    print("Ölüm oranı eşittir")
```
